```python
import random

def firstbatting():
    global GameOver
    player_runs = 0
    ai_runs = 0
    
    while not GameOver:
        player = int(input("Bat: "))
        ai = random.randint(1, 6)
        print("AI's number", ai)
        player_runs += player
        print("Your current score is", player_runs)
        
        if player == ai:
            print("You are out")
            print(f"AI needs {player_runs + 1} to win the game")
            while True:
                player = int(input("ball: "))
                ai = random.randint(1, 6)
                print("number of ai", ai)
                ai_runs += ai
                print("Current runs of AI is:", ai_runs)
                
                if ai_runs > player_runs:
                    print("AI won the game. Better luck next time!")
                    GameOver = True
                    return player_runs, ai_runs
                
                elif player == ai:
                    print("You got the wicket and won the match by", (player_runs - ai_runs, "runs"))
                    GameOver = True
                    return player_runs, ai_runs


def firstbowling():
    global GameOver
    ai_runs = 0
    player_runs = 0
    
    while not GameOver:
        player = int(input("Bowl: "))
        ai = random.randint(1, 6)
        print("AI batting:", ai)
        ai_runs += ai
        print("AI's current score is", ai_runs)
        
        while True:
            player = int(input("bat: "))
            ai = random.randint(1, 6)
            print("number of ai", ai)
            player_runs += ai
            print("Current runs of player is:", player_runs)
            
            if ai_runs > player_runs:
                print("AI won the game. Better luck next time!")
                GameOver = True
                return player_runs, ai_runs
                
            elif player == ai:
                print("You got the wicket and won the match by", (player_runs - ai_runs, "runs"))
                GameOver = True
                return player_runs, ai_runs


print("Welcome to the game!")
toss = input("Enter odd or even: ").lower()
player_num = int(input("Enter a number between 1 and 6: "))

if player_num > 6:
    print("Invalid input")
else:
    ai_num = random.randint(1, 6)
    print("AI's number", ai_num)
    GameOver = False

    total = player_num + ai_num
    if (total % 2 == 0 and toss == "even") or (total % 2 != 0 and toss == "odd"):
        print("You won the toss")
        choice = input("Choose 'bat' or 'bowl': ").lower()
        
        if choice == "bat":
            player_runs, ai_runs = firstbatting()
        elif choice == "bowl":
            ai_runs, player_runs = firstbowling()
        else:
            print("Invalid choice")
            
    else:
        print("AI won the toss")
        aichoice = random.choice(['bat', 'bowl'])
        
        if aichoice == "bat":
            ai_runs, player_runs = firstbowling()
        elif aichoice == "bowl":
            player_runs, ai_runs = firstbatting()
        else:
            print("Invalid choice")
```
